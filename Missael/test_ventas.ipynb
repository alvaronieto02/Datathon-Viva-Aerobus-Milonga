```python
product_types = df_ventas["ProductType"].unique()

type_dict = {}
type_df = {"Type":[],"List":[],"ItemCount":[]}

for typ in product_types:
    L = list( df_ventas["ProductName"][ df_ventas["ProductType"] == typ ].unique() )
    type_dict[typ] = L
    
    type_df["Type"].append(typ)
    type_df["List"].append(L)
    type_df["ItemCount"].append(len(L))
  
type_df = pd.DataFrame(type_df)
display(type_dict)
display(type_df)
```

```python
# Datos de ventas para entrenamiento de modelo. Usar el Flight_ID para pegar el resto de la información

df_ventas_sum = df_ventas[["Flight_ID","ProductType","Quantity"]].groupby(by=["Flight_ID","ProductType"], as_index=False).sum()
# df_ventas_id = df_ventas[["Flight_ID","ProductType"]].groupby(by="ProductType", as_index=False)

display(df_ventas_sum)
```

```python
df_vuelos_2 = df_vuelos.dropna()

# Factores que influencian la venta:
# - Factor de ocupación (OccupationFactor)
# - Origen / Destino del vuelo (Origin_Type, Destination_Type)
# - Duración del vuelo (FlightDuration)
# - Horario del vuelo (Time)
# - Temporada (Month)

df_vuelos_2["OccupationFactor"] = df_vuelos_2["Passengers"] / df_vuelos_2["Capacity"]
df_vuelos_2["FlightDuration"] = df_vuelos_2["STA"] - df_vuelos_2["STD"]
df_vuelos_2["Month"] = df_vuelos_2["STD"].dt.month
df_vuelos_2["Time"] = df_vuelos_2["STD"].dt.time

compras_p_vuelo = df_ventas[["Flight_ID","Quantity","TotalSales"]].groupby("Flight_ID").sum().reset_index()
compras_p_vuelo = pd.DataFrame(compras_p_vuelo)
map_ventas = dict(zip( compras_p_vuelo["Flight_ID"], compras_p_vuelo["TotalSales"] ))

df_vuelos_2["TotalSales"] = df_vuelos_2["Flight_ID"].map(map_ventas)

df_final = df_vuelos_2.dropna()

display(df_final)
```
