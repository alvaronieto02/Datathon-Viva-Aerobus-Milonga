# Exploracion de datos

```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

vuelos = 'C:\\Users\\max80\\OneDrive\\Documents\\DATATHON 2024\\Filghts TEC_Valid.csv'
compras = 'C:\\Users\\max80\\OneDrive\\Documents\\DATATHON 2024\\Sales TEC_Valid.csv'

df_vuelos = pd.read_csv(vuelos)
df_compras = pd.read_csv(compras)

n_vuelos = df_vuelos.shape[0]
n_compras = df_compras.shape[0]

vacios_vuelos = df_vuelos.isna()
vacios_vuelos = vacios_vuelos.sum()

vacios_compras = df_compras.isna()
vacios_compras = vacios_compras.sum()
```

```python
df_vuelos['STD'] = pd.to_datetime(df_vuelos['STD'])
df_vuelos['Año'] = df_vuelos['STD'].dt.year
df_vuelos['Mes'] = df_vuelos['STD'].dt.month

df_vuelos = df_vuelos.sort_values(by = 'STD', ascending = True)

pasajeros_mes = df_vuelos.groupby(['Año', 'Mes'])['Passengers'].sum()
```

# Creacion de la base de entrenamiento para modelo de pasajeros

```python
df_final = df_vuelos.dropna()

df_final['Departure'] = df_final['STD'].dt.time
df_final["Route"] = df_final["DepartureStation"] + "+" + df_final["ArrivalStation"]
df_final["Occupation fraction"] = df_final["Passengers"] / df_final["Capacity"]

df_final['STA'] = pd.to_datetime(df_final['STA'])
df_final['Arrival'] = df_final['STA'].dt.time

df_final['Duration'] = df_final['STA'] - df_final['STD']
```

# Asignacion de inputs y output

```python
categorias = df_final[['Route','Departure']]
numericas = df_final[['Mes','Capacity']]

encoder = LabelEncoder()  #Codificador de categorias
categorias = categorias.apply(encoder.fit_transform) #Codificacion

X = pd.concat([pd.DataFrame(numericas), categorias], axis=1) #Inputs array
Y = df_final['Occupation fraction']  #Output
```

# Modelo para prediccion de pasajeros

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=20)  #Particion de los datos de output en entrenamiento y prueba

model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),  
    tf.keras.layers.Dense(32, activation='relu'), 
    tf.keras.layers.Dense(16, activation='relu'),  
    tf.keras.layers.Dense(8, activation='relu'),   
    tf.keras.layers.Dense(4, activation='relu'),  
    tf.keras.layers.Dense(1)                      
])
model.compile(optimizer = 'adam', loss = 'mean_squared_error')   #Compilador del modelo, optimizador y funcion de error
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))  #Entrenamiento
```

# Prueba de errores con datos conocidos

```python
prediccion_23 = model.predict(X)
prediccion_23 = prediccion_23.flatten()
df_final['Neurona'] = prediccion_23
```

```python
df_final['ERROR'] = (df_final['Occupation fraction'] - df_final['Neurona'])**2
error_medio = df_final['ERROR'].mean()
df_final['Passengers Prediction'] = df_final['Neurona'] * df_final['Capacity']
display(df_final)

print("El error promedio entre las predicciones es del:",error_medio*100,"%")
```

# Predicción de los pasajeros para 2024

```python
df_predict = df_vuelos.loc[(df_vuelos['Año'] == 2024) | (df_vuelos['Año'] == 2025)]
df_predict['Departure'] = df_predict['STD'].dt.time
df_predict["Route"] = df_predict["DepartureStation"] + "+" + df_predict["ArrivalStation"]

df_predict['STA'] = pd.to_datetime(df_predict['STA'])
df_predict['Arrival'] = df_predict['STA'].dt.time

df_predict['Duration'] = df_predict['STA'] - df_predict['STD']

categorias_pre = df_predict[['Route','Departure']]
numericas_pre = df_predict[['Mes','Capacity']]

encoder = LabelEncoder()
categorias_pre = categorias_pre.apply(encoder.fit_transform) #Codificacion

X_pre = pd.concat([pd.DataFrame(numericas_pre), categorias_pre], axis=1) #Inputs prediccion
prediccion_pasajeros = model.predict(X_pre)
prediccion_pasajeros = prediccion_pasajeros.flatten()
df_predict['Prediccion de NN'] = prediccion_pasajeros
display(df_predict)
```

```python
df_predict['Passengers'] = df_predict['Prediccion de NN']*df_predict['Capacity']
```

# Asignación de inputs y outputs para cantidad de productos

```python
df_compras_sum = df_compras[["Flight_ID","ProductType","Quantity"]].groupby(by=["Flight_ID","ProductType"], as_index=False).sum()
df_compras_sum = pd.DataFrame(df_compras_sum)

mapeo_passengers = dict(zip(df_final["Flight_ID"], df_final["Passengers"]))
df_compras_sum["Passengers"] = df_compras_sum["Flight_ID"].map(mapeo_passengers)

mapeo_route = dict(zip(df_final["Flight_ID"], df_final["Route"]))
df_compras_sum["Route"] = df_compras_sum["Flight_ID"].map(mapeo_route)

mapeo_mes = dict(zip(df_final["Flight_ID"], df_final["Mes"]))
df_compras_sum["Mes"] = df_compras_sum["Flight_ID"].map(mapeo_mes)

mapeo_duration = dict(zip(df_final["Flight_ID"], df_final["Duration"]))
df_compras_sum["Duration"] = df_compras_sum["Flight_ID"].map(mapeo_duration)

df_compras_sum = df_compras_sum.dropna(subset=['Mes'])
df_compras_sum = df_compras_sum.dropna(subset=['Passengers'])
df_compras_sum = df_compras_sum.dropna(subset=['Duration'])
df_compras_sum = df_compras_sum.astype({'Mes': 'int64','Passengers': 'int64'})
```

```python
categorys_products_2023 = df_compras_sum[['Route','ProductType','Duration']]
numerics_products_2023 = df_compras_sum[['Mes','Passengers']]

categorys_products_2023 = categorys_products_2023.apply(encoder.fit_transform)

X_products_2023 = pd.concat([pd.DataFrame(numerics_products_2023), categorys_products_2023], axis=1) #Inputs array
Y_products_2023 = df_compras_sum['Quantity']  #Output
```

# Modelo 2

```python
X_train_product, X_test_product, y_train_product, y_test_product = train_test_split(X_products_2023, Y_products_2023, test_size=0.2, random_state=20)  #Particion de los datos de output en entrenamiento y prueba

model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(80, activation='relu', input_shape=(X_train_product.shape[1],)),  
    tf.keras.layers.Dense(40, activation='relu'), 
    tf.keras.layers.Dense(20, activation='relu'),  
    tf.keras.layers.Dense(10, activation='relu'),   
    tf.keras.layers.Dense(5, activation='relu'),  
    tf.keras.layers.Dense(1)                      
])
model_2.compile(optimizer = 'adam', loss = 'mean_squared_error')   #Compilador del modelo, optimizador y funcion de error
model_2.fit(X_train_product, y_train_product, epochs=20, batch_size=32, validation_data=(X_test_product, y_test_product))
```

```python
prediccion_products_2023 = model_2.predict(X_products_2023)
prediccion_products_2023 = prediccion_products_2023.flatten()
df_compras_sum['Prediccion de NN'] = prediccion_products_2023
```
