```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

# Flights Dataframe


```python
# Convert .csv file into dataframe
flights_path = 'C:\\Users\\User\\Desktop\\Filghts TEC_Valid.csv'
flights_df = pd.read_csv(flights_path)

# Convert departure and arrival dates to datetime format
flights_df['STD'] = pd.to_datetime(flights_df['STD'])
flights_df['STA'] = pd.to_datetime(flights_df['STA'])

# Sort flights by departure date
flights_df = flights_df.sort_values(by='STD')

# Insert flight duration column
flights_df.insert(flights_df.columns.get_loc('Capacity'), 'Duration',
                  ((flights_df['STA'] - flights_df['STD']).dt.total_seconds() / 60).astype(int))

# Initialize occupancy column
flights_df['Occupancy'] = np.nan
```

```python
# Separate dataframe for 2023 and future flights
flights_2023 = flights_df[flights_df['STD'] < '2024-01-01'].copy()
flights_future = flights_df[flights_df['STD'] >= '2024-01-01'].copy()

# Convert 2023 passengers and bookings into integers
flights_2023['Passengers'] = flights_2023['Passengers'].astype(int)
flights_2023['Bookings'] = flights_2023['Bookings'].astype(int)

# Insert 2023 flights occupancy
flights_2023['Occupancy'] = flights_2023['Passengers'] / flights_2023['Capacity']

# Display 2023 flights dataframe
display(flights_2023)
```

```python
print(f"No. of flights in 2023: {flights_2023.shape[0]}")
```
