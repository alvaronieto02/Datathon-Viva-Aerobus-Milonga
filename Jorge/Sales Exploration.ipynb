# Sales Dataframe

```python
# Display sales dataframe
sales_path = 'C:\\Users\\User\\Desktop\\Sales TEC_Valid.csv'
sales_df = pd.read_csv(sales_path)

display(sales_df)
```

```python
# Type of products sold
product_types = sorted(sales_df['ProductType'].unique())

print(product_types)

```

```python
grouped = sales_df.groupby('ProductType')
products = {}

# Iterate over the groups
for product_type, group in grouped:
    products[product_type] = group['ProductName'].unique()

    print(product_type)
    print(products[product_type])
    print()
```
